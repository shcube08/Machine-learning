## Line fitting using Maximum Likelihood Function

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import norm
```

```python
x = np.array([1, 2, 3])
y = np.array([1.2, 1.9, 3.2])
n = len(x)
print(x)
print(y)
print(n)
```

### Plotting given x and y

```python
plt.plot(x, y)
plt.scatter(x, y, color='black')
plt.show()
```

```python
def MLE(x, y, n):
    sigma_xy = x.dot(y.T)
    sigma_x = np.sum(x)
    sigma_y = np.sum(y)
    sigma_x2 = x.dot(x.T)
    
    w1 = ((n*sigma_xy) - (sigma_x*sigma_y)) / ((n*sigma_x2) - (sigma_x**2))
    w0 = ((sigma_y*sigma_x2) - (sigma_xy*sigma_x)) / ((n*sigma_x2) - (sigma_x**2))
    return w0, w1
```

```python
w0, w1 = MLE(x, y, n)
print(f"Intercept: {w0}, Slope: {w1}")
```

```python
def getEstimatedModel(w0, w1, n):
    y_hat = []
    for i in range(n):
        value = w0 + w1*x[i]
        y_hat.append(value)
    return np.array(y_hat)
```

```python
y_hat = getEstimatedModel(w0, w1, n)
print(y_hat)
```

### Plotting the estimated model x and y_hat

```python
plt.scatter(x=x, y=y_hat, color='black')
plt.plot(x, y_hat)
plt.show()
```

### Plotting estimated model over the actutal values y

```python
plt.scatter(x, y, color='black')
plt.plot(x, y_hat)
plt.xlabel('X')
plt.ylabel('y_hat')
plt.title('Estimated Model')
plt.show()
```
